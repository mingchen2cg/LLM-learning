激活函数（Activation Function）是深度学习模型的灵魂，它们决定了神经网络如何处理非线性特征。

以下是对 **Sigmoid**、**ReLU**、**SiLU** 以及其他主流激活函数的全面总结与对比。

---

## 1. 经典激活函数：Sigmoid & Tanh

这些函数在早期神经网络（如 MLP, RNN）中非常流行，但现在更多用于特定场景（如输出层或门控机制）。

### Sigmoid
* **公式**：$\sigma(x) = \frac{1}{1 + e^{-x}}$
* **输出范围**：$(0, 1)$
* **优点**：平滑，输出可解释为概率。
* **缺点**：
    * **梯度消失**：当 $x$ 很大或很小时，梯度趋于 0。
    * **非零中心化**：输出全为正，会导致权重更新时的震荡。
    * **计算量大**：涉及指数运算。

### Tanh (双曲正切)
* **公式**：$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$
* **输出范围**：$(-1, 1)$
* **优点**：**零中心化**（Zero-centered），收敛速度通常快于 Sigmoid。
* **缺点**：依然存在梯度消失问题。



---

## 2. 现代主流：ReLU 家族

ReLU 的出现彻底改变了深度学习，解决了深度网络训练难的问题。

### ReLU (Rectified Linear Unit)
* **公式**：$f(x) = \max(0, x)$
* **优点**：
    * **计算极其简单**：只需判断阈值。
    * **缓解梯度消失**：正区间梯度恒为 1。
    * **稀疏性**：部分神经元被抑制，提高泛化能力。
* **缺点**：**Dead ReLU 问题**。如果权重更新步长过大，神经元可能陷入“永久沉默”（梯度恒为 0）。

### Leaky ReLU / PReLU
* **公式**：$f(x) = \max(\alpha x, x)$ （$\alpha$ 通常为 0.01）
* **目的**：给负区间留一点微小的梯度，防止神经元“死掉”。



---

## 3. 大模型与 Transformer 的新宠：Swish/SiLU & GELU

随着深度学习进入 Transformer 时代，具有**平滑性**和**非单调性**的激活函数逐渐取代了 ReLU。

### SiLU (Sigmoid Linear Unit) / Swish
* **公式**：$f(x) = x \cdot \sigma(x) = \frac{x}{1 + e^{-x}}$
* **特点**：
    * **无上界、有下界**。
    * **非单调性**：在 $x < 0$ 的部分区域，函数值会先降后升。
    * **平滑性**：在 $x=0$ 处连续且可导，实验证明在深度模型（如 **YOLOv5/v8**, **Llama**）中效果优于 ReLU。

### GELU (Gaussian Error Linear Unit)
* **公式**：$f(x) \approx 0.5x(1 + \tanh[\sqrt{2/\pi}(x + 0.044715x^3)])$
* **背景**：BERT、GPT、ViT 等 Transformer 模型的**标配**。
* **逻辑**：它结合了 ReLU 的确定性和 Dropout 的随机性，根据输入的幅值大小决定信息通过的概率。



---

## 4. 激活函数综合对比表

| 激活函数 | 公式定义 | 优势 | 劣势 | 常见应用场景 |
| :--- | :--- | :--- | :--- | :--- |
| **Sigmoid** | $1/(1+e^{-x})$ | 平滑，概率输出 | 梯度消失，计算重 | 二分类输出层，门控 |
| **Tanh** | $\tanh(x)$ | 零中心化 | 梯度消失 | 循环神经网络 (RNN) |
| **ReLU** | $\max(0, x)$ | 计算极快，缓解梯度消失 | Dead ReLU 问题 | 卷积神经网络 (CNN) |
| **Leaky ReLU** | $\max(\alpha x, x)$ | 解决神经元死亡 | $\alpha$ 超参数难调 | 缓解 ReLU 训练困难 |
| **SiLU** | $x \cdot \sigma(x)$ | 平滑、非单调，精度高 | 计算量略大于 ReLU | YOLO 系列, Llama |
| **GELU** | $x \cdot \Phi(x)$ | 符合随机正则化思想 | 计算复杂 | Transformer (GPT, BERT) |

---

## 总结与建议

1.  **入门/基础 CNN**：首选 **ReLU**，如果训练不收敛，试试 **Leaky ReLU**。
2.  **Transformer/大语言模型**：无脑选择 **GELU** 或 **SiLU (Swish)**。
3.  **循环神经网络 (RNN/LSTM)**：隐藏层通常用 **Tanh**。
4.  **分类任务末端**：二分类用 **Sigmoid**，多分类用 **Softmax**。

现在的趋势是：激活函数越来越趋向于**平滑**和**在 0 点附近允许微小的负值**，这有助于捕捉更复杂的非线性关系并加速收敛。
