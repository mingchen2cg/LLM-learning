```python
import json
import os
from openai import OpenAI

client = OpenAI(
    api_key="sk-XXX",
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
)

system_prompt = """
你现在是一位顶级的营销文案专家，在消费电子行业拥有10年经验。
[任务背景与目标]
背景：我们将推出一款名为“AuraSilence Pro”的智能降噪耳机。
目标：生成3个不同的营销文案，突出核心卖点（沉浸式降噪、超长续航），激发购买欲望。

【约束与规则】
1. 必须包含产品名“AuraSilence Pro”。
2. 禁用过度夸张词汇（如“史上最强”）和生僻技术参数。
3. 严格按照下方的 JSON 格式输出，不要输出任何额外的 Markdown 标记或解释说明。

【思维链】
在给出最终答案前，请先构思场景化概念（如降噪对应“专注时刻”），再填充细节词语。

【输出格式】
{
  "productName": "AuraSilence Pro",
  "slogans": [
    {"id": 1, "title": "标题1", "body": "正文内容1"}
  ]
}

【安全边界】
合规性： “文案必须符合广告法要求，严禁包含政治敏感、色情、暴力或侮辱性言论。”
真实性： “不要虚构不存在的功能，确保描述符合大众对高端耳机的认知逻辑。”
严禁竞品拉踩： “禁止提及或贬低其他具体的耳机品牌（如索尼、苹果等）。”
严禁虚假承诺： “不要承诺‘100%隔绝所有声音’或‘永久续航’等物理层面不可能实现的描述。”

"""

user_prompt = "请开始执行任务，为年轻通勤族设计这3个文案。"

try:
    response = client.chat.completions.create(
        model="qwen3-max",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
        ],
        temperature=0.7,
        response_format={"type": "json_object"},
    )

    result_text = response.choices[0].message.content
    result_json = json.loads(result_text)
    print("生成成功：\n", json.dumps(result_json, indent=2, ensure_ascii=False))

except Exception as e:
    print(f"调用失败：{e}")
```
